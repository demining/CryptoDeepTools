```
#@title Parameters
#@markdown List of address patterns (comma-separated)
patterns = "1MyPattern" #@param {type: "string"}
#@markdown Verbose output
verbose = False #@param {type: "boolean"}
#@markdown Quiet output
quiet = False #@param {type: "boolean"}
#@markdown Simulate
simulate = False #@param {type: "boolean"}
#@markdown Use regular expression match instead of prefix
regex = False #@param {type: "boolean"}
#@markdown Keep pattern and continue search after finding a match
keep_searching = False #@param {type: "boolean"}
#@markdown Case-insensitive prefix search
case_insensitive = False #@param {type: "boolean"}
#@markdown Stop after first match
stop_after_first = False #@param {type: "boolean"}
#@markdown Stop after generating <amount> patterns
stop_after = 5 #@param {type: "integer"}
#@markdown Coin type
coin = "BTC" #@param BTC ["42", "AC", "AIB", "ANC", "ARS", "ATMOS", "AUR", "AXE", "BLAST", "BLK", "BWK", "BQC", "BTC", "TEST", "BTCD", "CCC", "CCN", "CDN", "CLAM", "CNC", "CNOTE", "CON", "CRW", "DASH", "DEEPONION", "DNR", "DGB", "DGC", "DMD", "DOGED", "DOGE", "DOPE", "DVC", "EFL", "EMC", "FAIR", "FLOZ", "FTC", "GAME", "GAP", "GCR", "GRC", "GRLC", "GRS", "GUN", "HAM", "HBN", "HODL", "IXC", "JBS", "JIN", "LBRY", "LEAF", "LTC", "MMC", "MONA", "MUE", "MYRIAD", "MZC", "NEET", "NEOS", "NLG", "NMC", "NVC", "NYAN", "OK", "OMC", "PIGGY", "PINK", "PIVX", "PKB", "PND", "POT", "PPC", "PTC", "PTS", "QTUM", "RBY", "RDD", "RIC", "ROI", "SCA", "SDC", "SKC", "SPR", "START", "SXC", "TPC", "UIS", "UNO", "VIA", "VPN", "VTC", "WDC", "WKC", "WUBS", "XC", "XPM", "YAC", "YTN", "ZNY", "ZOOM", "ZRC"]
#@markdown Generate address with the given version
pubkey_version = "" #@param {type: "string"}
#@markdown Specify private key version (pubkey_version provides public key)
privkey_version = "" #@param {type: "string"}
#@markdown Generate address with the given format
format = "pubkey" #@param pubkey ["pubkey", "compressed", "script"]
#@markdown Specify base public key for priecewise key generation
pubkey = "" #@param {type: "string"}
#@markdown Encrypt private keys, prompt for password
encrypt_keys = False #@param {type: "boolean"}
#@markdown Encrypt private keys with <password> (UNSAFE)
encrypt_keys_param = "" #@param {type: "string"}
#@markdown Set number of worker threads (Default: number of CPUs)
threads =  1024#@param {type: "integer"}
#@markdown File containing list of patterns, one per line
in_file = "" #@param {type: "string"}
#@markdown Write pattern matches to <file>
out_file = "" #@param {type: "string"}
#@markdown Seed random number generator from <file>
seed_file = "" #@param {type: "string"}
#@markdown Private key prefix in hex (vipco.in)
privkey_prefix = "" #@param {type: "string"}
#@markdown Format output of matches in CSV (disables verbose mode)
#@markdown Output as [COIN],[PREFIX],[ADDRESS],[PRIVKEY]
csv = False #@param {type: "boolean"}

#CLI arguments processing
from tensorflow.test import gpu_device_name
param = [
    "./" + ("ocl" if gpu_device_name() == "/device:GPU:0" else "") + "vanitygen", 
    "-C", coin, "-F", format
]
if verbose:
  param.append("-v")
if quiet:
  param.append("-q")
if simulate:
  param.append("-s")
if regex:
  param.append("-r")
if case_insensitive:
  param.append("-r")
if keep_searching:
  param.append("-k")
if stop_after_first:
  param.append("-1")
if stop_after:
  param.append("-a")
  param.append(str(stop_after))
if pubkey_version:
  param.append("-X")
  param.append(pubkey_version)
if privkey_version:
  param.append("-Y")
  param.append(privkey_version)
if pubkey:
  param.append("-P")
  param.append(pubkey)
if encrypt_keys:
  param.append("-e")
if encrypt_keys_param:
  param.append("-E")
  param.append(encrypt_keys_param)
if threads:
  param.append("-t")
  param.append(str(threads))
if in_file:
  param.append("-f")
  param.append(in_file)
if out_file:
  param.append("-o")
  param.append(out_file)
if seed_file:
  param.append("-s")
  param.append(seed_file)
if privkey_prefix:
  param.append("-Z")
  param.append(privkey_prefix)
if csv:
  param.append("-z")
  param.append(csv)
for pattern in patterns.split(","):
    param.append(pattern.replace("\\", "\\\\").replace("\"", "\\\""))
```

```
#@title Run
#@markdown After you have configured VanityGen parameters, run this step 
#@markdown to generate keys.

try:
  %pushd "linux-binary"
  from shlex import quote
  cli = " ".join(map(quote, param))
  !$cli
finally:
  %popd
```
